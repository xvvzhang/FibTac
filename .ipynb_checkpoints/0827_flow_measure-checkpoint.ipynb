```python
import os
import numpy as np
import torch
from torch.utils.data import Dataset
import kornia.augmentation as K
from kornia.augmentation import VideoSequential
import cv2
from glob import glob
import random
import pandas as pd
import re

def yellow_dot_mask(img_bgr,
                    hsv_lower=(18, 60, 80),   # H,S,V lower bound (OpenCV H:0..179)
                    hsv_upper=(60, 255, 255), # H,S,V upper bound
                    min_area=5,               # remove blobs smaller than this
                    blur_ksize=5):
    """
    Binary mask of yellow dots from a BGR image.
    1) blur -> HSV threshold (loose yellow)
    2) equalize V channel + Otsu threshold
    3) AND the two masks
    4) morphology + small object removal
    """
    if blur_ksize and blur_ksize > 1:
        img_bgr = cv2.GaussianBlur(img_bgr, (blur_ksize, blur_ksize), 0)

    hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

    # HSV color mask
    lower = np.array(hsv_lower, dtype=np.uint8)
    upper = np.array(hsv_upper, dtype=np.uint8)
    mask_hsv = cv2.inRange(hsv, lower, upper)

    # Brightness refinement on V channel (equalize + Otsu)
    v = hsv[..., 2]
    v_eq = cv2.equalizeHist(v)
    _, mask_v = cv2.threshold(v_eq, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Combine color + brightness constraints
    mask = cv2.bitwise_and(mask_hsv, mask_v)

    # Morphology cleanup
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=1)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=1)

    # Remove tiny components
    num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(mask, connectivity=8)
    clean = np.zeros_like(mask)
    for i in range(1, num_labels):  # 0 is background
        if stats[i, cv2.CC_STAT_AREA] >= min_area:
            clean[labels == i] = 255
    return clean


class RegressionDataset(Dataset):
    """
    Folder layout:
      root/
        2_8/
          1/                 <-- instance
            000001.jpg
            ...
          2/
        1_14/
          ...

    Target per instance:
      [angle, speed] where:
        angle = first number in top folder name (e.g., 2 in '2_8')
        level = second number in top folder name (e.g., 8 in '2_8')
        speed = lookup(level, angle) from Excel
    """
    def __init__(self, root_dir, xlsx_path, train=True, device='cuda', max_frames=300):
        self.root_dir = root_dir
        self.device = torch.device(device if torch.cuda.is_available() else 'cpu')
        self._train = bool(train)
        self.max_frames = int(max_frames)

        # Build lookup (level, angle) -> speed
        self.speed_lookup = self._build_speed_lookup(xlsx_path)

        # Storage aligned by index
        self.paths = []       # instance folder paths
        self.targets = []     # (angle, speed) float tuples
        self.data = []        # np arrays [T, 1, 224, 224]

        self._set_runtime_transform()
        self._index_folders()
        self._preload_all()

    # ---- train toggle ----
    @property
    def train(self):
        return self._train

    @train.setter
    def train(self, value: bool):
        self._train = bool(value)
        self._set_runtime_transform()

    # ---- normalization exactly like your original (single-channel) ----
    def _set_runtime_transform(self):
        self.transform = VideoSequential(
            K.Normalize(mean=torch.tensor([0.5], device=self.device),
                        std=torch.tensor([0.5], device=self.device)),
            same_on_frame=True,
        ).to(self.device)

    # ---- Excel -> dict[(level, angle)] = speed ----
    @staticmethod
    def _build_speed_lookup(xlsx_path: str):
        df = pd.read_excel(xlsx_path, engine="openpyxl")
        first_col = df.columns[0]
        df = df.rename(columns={first_col: "Level"}).set_index("Level")
        lookup = {}
        for level in df.index:
            for col in df.columns:
                try:
                    angle = int(str(col).strip())
                except ValueError:
                    continue
                lookup[(int(level), int(angle))] = float(df.loc[level, col])
        return lookup

    # ---- index: parse top folder "angle_level" ----
    def _index_folders(self):
        pat = re.compile(r'^(\d+)[_-](\d+)$')  # accept 2_8 or 2-8

        for top in sorted(os.listdir(self.root_dir)):
            top_path = os.path.join(self.root_dir, top)
            if not os.path.isdir(top_path) or top.startswith('.'):
                continue

            m = pat.match(top)
            if not m:
                continue
            angle = int(m.group(1))
            level = int(m.group(2))

            key = (level, angle)
            if key not in self.speed_lookup:
                raise ValueError(f"No speed for (level={level}, angle={angle}) in Excel.")

            speed = self.speed_lookup[key]

            for inst in sorted(os.listdir(top_path)):
                inst_path = os.path.join(top_path, inst)
                if not os.path.isdir(inst_path) or inst.startswith('.'):
                    continue
                self.paths.append(inst_path)
                self.targets.append((float(angle), float(speed)))

    # ---- helpers ----
    @staticmethod
    def _center_crop(img: np.ndarray, target_h: int, target_w: int) -> np.ndarray:
        h, w = img.shape[:2]
        ch = min(target_h, h)
        cw = min(target_w, w)
        y0 = (h - ch) // 2
        x0 = (w - cw) // 2
        return img[y0:y0+ch, x0:x0+cw]

    def _read_and_preprocess_frames(self, folder_path: str) -> np.ndarray:
        """
        Deterministic preload (single-channel pipeline, same as yours):
          - Read COLOR (BGR) so your yellow_dot_mask can work
          - Apply yellow_dot_mask (if available)
          - Center-crop to 480x480
          - Resize to 224x224
          - Convert to GRAYSCALE (single channel)
          - Scale to [0,1] and add channel dim -> [1, 224, 224]
        Returns np.float32 array of shape [T, 1, 224, 224].
        """
        frame_paths = sorted(glob(os.path.join(folder_path, '*.jpg')))
        frames = []
        start_idx = 0

        for path in frame_paths[start_idx:]:
            img_bgr = cv2.imread(path, cv2.IMREAD_COLOR)
            if img_bgr is None:
                continue

            img_bgr = self._center_crop(img_bgr, 300, 300)
            img_bgr = cv2.resize(img_bgr, (224, 224), interpolation=cv2.INTER_LINEAR)

            img = yellow_dot_mask(img_bgr)

            # convert to single channel (keep your dataset's C=1 convention)
            #img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)  # [224,224]

            # [0,1] and add channel dimension -> [1,224,224]
            img = (img.astype(np.float32) / 255.0)[None, ...]
            frames.append(img)

            if len(frames) >= self.max_frames:
                break

        if not frames:
            print(f"[WARNING] No valid frames in: {folder_path}")
            return np.zeros((1, 1, 224, 224), dtype=np.float32)

        return np.stack(frames, axis=0)  # [T, 1, 224, 224]

    def _preload_all(self):
        self.data = []
        for p in self.paths:
            self.data.append(self._read_and_preprocess_frames(p))

    # ---- Dataset API ----
    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        frames_np = self.data[idx]          # [T, 1, 224, 224]
        angle, speed = self.targets[idx]
        folder = self.paths[idx]

        # Same even/odd thinning behavior
        if self._train:
            choice = torch.randint(0, 2, ()).item()
            frames_np = frames_np[choice::2]
        else:
            frames_np = frames_np[0::2]

        # -> tensor on device; VideoSequential expects [B,T,C,H,W] (C=1)
        frames = torch.from_numpy(frames_np).to(self.device, non_blocking=True)  # [T,1,224,224]
        frames = frames.unsqueeze(0)                                             # [1,T,1,224,224]

        # Normalize (single-channel)
        frames = self.transform(frames)                                          # [1,T,1,224,224]
        frames = frames.squeeze(0)                                               # [T,1,224,224]

        #target = torch.tensor([angle, speed], dtype=torch.float32, device=self.device)


        # 改成 min-max 归一化
        angle_min, angle_max = 1.0, 7.0
        speed_min, speed_max = 9.7, 16.6

        angle_norm = (angle - angle_min) / (angle_max - angle_min)
        speed_norm = (speed - speed_min) / (speed_max - speed_min)

        target = torch.tensor([angle_norm, speed_norm], dtype=torch.float32, device=self.device)

        return frames, target, folder
```

```python
import random
from collections import defaultdict
from torch.utils.data import Subset
from torch.utils.data import random_split, DataLoader

def stratified_split(dataset_class, root_dir, split_ratio=0.8, seed=42, **dataset_kwargs):
    """
    Perform a stratified split for a dataset based on its labels.

    Args:
        dataset_class: Dataset class to instantiate (must have a 'labels' attribute).
        root_dir (str): Root directory passed to the dataset class.
        split_ratio (float): Proportion of data for training.
        seed (int): Random seed for reproducibility.
        **dataset_kwargs: Extra keyword arguments passed to dataset_class.

    Returns:
        (train_subset, val_subset): Subsets for training and validation.
    """
    # Load full dataset (assuming it can expose labels)
    full_dataset = dataset_class(root_dir=root_dir, train=True, **dataset_kwargs)

    # Group indices by class
    label_to_indices = defaultdict(list)
    for idx, label in enumerate(full_dataset.labels):
        label_to_indices[label].append(idx)

    random.seed(seed)

    train_indices, val_indices = [], []
    for label, indices in label_to_indices.items():
        random.shuffle(indices)
        split_point = int(split_ratio * len(indices))
        train_indices.extend(indices[:split_point])
        val_indices.extend(indices[split_point:])

    # Create train and val datasets
    train_dataset = dataset_class(root_dir=root_dir, train=True, **dataset_kwargs)
    val_dataset   = dataset_class(root_dir=root_dir, train=False, **dataset_kwargs)

    train_subset = Subset(train_dataset, train_indices)
    val_subset   = Subset(val_dataset, val_indices)

    return train_subset, val_subset


train_dataset, val_dataset = stratified_split(
    dataset_class=ClassDataset,
    root_dir='./',
    split_ratio=1
)
train_dataloader = DataLoader(train_dataset, batch_size=8, shuffle=True, drop_last=True)
val_dataloader = DataLoader(val_dataset, batch_size=8, shuffle=False)
print(train_dataset.dataset.train)
print(val_dataset.dataset.train)
```

```python
import torchvision.models as models
import torch
import torch.nn as nn

class CRNNRegressor(nn.Module):
    def __init__(self):
        super(CRNNRegressor, self).__init__()

        self.resnet = models.resnet18(weights=None)
        self.resnet.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
        modules = list(self.resnet.children())[:-1]  # Remove avgpool and fc
        self.resnet = nn.Sequential(*modules)

        self.lstm = nn.LSTM(input_size=512, hidden_size=128, batch_first=True)

        # Output 2 values: [angle, speed]
        self.regressor = nn.Sequential(
            nn.Linear(128, 32),
            nn.ReLU(),
            nn.Linear(32, 2)
        )

    def forward(self, x):
        B, T, C, H, W = x.size()
        x = x.view(B * T, C, H, W)

        features = self.resnet(x)  # (B*T, 512, 1, 1)
        features = features.view(B, T, -1)  # (B, T, 512)

        lstm_out, _ = self.lstm(features)  # (B, T, 128)
        last_timestep = lstm_out[:, -1, :]  # (B, 128)

        out = self.regressor(last_timestep)  # (B, 2)
        return out
```

```python
import torchvision.models as models

import torch
import torch.nn as nn
from torchvision import models

class CRNN(nn.Module):
    def __init__(self, num_classes=4):
        super(CRNN, self).__init__()

        # Load a ResNet model (without pre-trained weights, customized for 1 channel input)
        self.resnet = models.resnet18(weights=None)
        # Modify the first convolutional layer to accept 1-channel inputs (112x112 input size)
        self.resnet.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
        modules = list(self.resnet.children())[:-1]  # Remove the last fully connected layer and global average pooling
        self.resnet = nn.Sequential(*modules)

        # LSTM for the combined features
        self.lstm = nn.LSTM(512, 128, batch_first=True)  # LSTM input: 512 (ResNet output size), hidden size: 128

        # Final fully connected layer for classification
        self.mlp2 = nn.Sequential(
            nn.Linear(128, 32),
            
            nn.ReLU(),
            nn.Linear(32, num_classes)  # Output layer for 4 classes
        )

    def forward(self, x):
        # Input x is of shape (batch_size, timesteps, C, H, W) -> (batch_size, 60, 1, 112, 112)
        batch_size, timesteps, C, H, W = x.size()
        
        # Flatten batch size and timesteps to pass through CNN
        cnn_input = x.view(batch_size * timesteps, C, H, W)
        
        # Pass input through CNN (ResNet backbone)
        features = self.resnet(cnn_input)  # Output shape: (batch_size * timesteps, 512, 1, 1)
        
        # Reshape features for LSTM input: (batch_size, timesteps, 512)
        features = features.view(batch_size, timesteps, -1)
        
        # Pass features through LSTM
        output, (hn, cn) = self.lstm(features)  # Output shape: (batch_size, timesteps, 128)

        # Classify the last step of LSTM output (only use the last timestep)
        output = self.mlp2(output[:, -1, :])  # Output shape: (batch_size, num_classes)
        return output


# 创建模型
num_classes = 6  # 替换为你的类别数
model = CRNN(num_classes=num_classes)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = nn.DataParallel(model)
model = model.to(device)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import math

# -----------------------------
# Model: use your regressor
# -----------------------------
# If your CRNNRegressor's first conv is set to C=1, keep as-is.
# If you kept it at 3 channels, set EXPAND_TO_3C=True to repeat grayscale.
# from your_model_file import CRNNRegressor  # <- adjust import

EXPAND_TO_3C = False  # set True ONLY if model expects 3-channel input

device = 'cuda' if torch.cuda.is_available() else 'cpu'
device_type = 'cuda' if device == 'cuda' else 'cpu'

model = CRNNRegressor()
model = nn.DataParallel(model)
model = model.to(device)

if device == 'cuda':
    torch.backends.cudnn.benchmark = True

# -----------------------------
# Metrics (validation/test)
# -----------------------------
@torch.no_grad()
def evaluate_regression(dataloader, model, device, use_amp: bool):
    model.eval()
    sum_mae_angle = 0.0
    sum_mae_speed = 0.0
    sum_mse_angle = 0.0
    sum_mse_speed = 0.0
    n = 0

    with torch.autocast(device_type=device_type, enabled=use_amp):
        for data, target, *_ in dataloader:
            data = data.to(device, non_blocking=(device == 'cuda'))
            target = target.to(device, non_blocking=(device == 'cuda'))

            if EXPAND_TO_3C and data.size(2) == 1:
                data = data.repeat(1, 1, 3, 1, 1)

            pred = model(data)
            diff = pred - target

            mae = diff.abs()
            mse = diff.pow(2)

            sum_mae_angle += mae[:, 0].sum().item()
            sum_mae_speed += mae[:, 1].sum().item()
            sum_mse_angle += mse[:, 0].sum().item()
            sum_mse_speed += mse[:, 1].sum().item()
            n += target.size(0)

    model.train()
    if n == 0:
        return {
            "mae_angle": 0.0, "mae_speed": 0.0, "mae_mean": 0.0,
            "rmse_angle": 0.0, "rmse_speed": 0.0, "rmse_mean": 0.0
        }

    mae_angle = sum_mae_angle / n
    mae_speed = sum_mae_speed / n
    rmse_angle = math.sqrt(sum_mse_angle / n)
    rmse_speed = math.sqrt(sum_mse_speed / n)
    return {
        "mae_angle": mae_angle,
        "mae_speed": mae_speed,
        "mae_mean": 0.5 * (mae_angle + mae_speed),
        "rmse_angle": rmse_angle,
        "rmse_speed": rmse_speed,
        "rmse_mean": 0.5 * (rmse_angle + rmse_speed),
    }

# -----------------------------
# Loss / optimizer / AMP
# -----------------------------
criterion = nn.SmoothL1Loss()  # Huber (L1-smooth) loss
#optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
optimizer = optim.AdamW(model.parameters(), lr=1e-5)

use_amp = (device == 'cuda')
scaler = torch.cuda.amp.GradScaler(enabled=use_amp)

# Track best checkpoints
best_val_mae = float('inf')
best_test_mae = float('inf')

best_val_path  = 'best_regressor_val.pth'
best_test_path = 'best_regressor_test.pth'
save_every_epoch_path = 'model_epoch{}.pth'

# -----------------------------
# Training loop
# -----------------------------
num_epochs = 200
log_every = 1
grad_clip = 1.0

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    batch_count = 0

    for data, target, *_ in train_dataloader:
        data = data.to(device, non_blocking=(device == 'cuda'))
        target = target.to(device, non_blocking=(device == 'cuda'))

        if EXPAND_TO_3C and data.size(2) == 1:
            data = data.repeat(1, 1, 3, 1, 1)

        optimizer.zero_grad(set_to_none=True)

        with torch.autocast(device_type=device_type, enabled=use_amp):
            pred = model(data)
            loss = criterion(pred, target)

        if use_amp:
            scaler.scale(loss).backward()
            if grad_clip is not None:
                scaler.unscale_(optimizer)
                nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
            scaler.step(optimizer)
            scaler.update()
        else:
            loss.backward()
            if grad_clip is not None:
                nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
            optimizer.step()

        running_loss += loss.item()
        batch_count += 1

    # ---- end epoch: evaluate ----
    train_loss = running_loss / max(1, batch_count)
    val_metrics  = evaluate_regression(val_dataloader,  model, device, use_amp)
    test_metrics = evaluate_regression(test_dataloader, model, device, use_amp)

    val_mae_mean  = val_metrics["mae_mean"]
    test_mae_mean = test_metrics["mae_mean"]

    if (epoch + 1) % log_every == 0:
        print(
            f"Epoch [{epoch+1}/{num_epochs}] "
            f"TrainLoss: {train_loss:.4f} | "
            f"Val MAE (ang/spd/mean): {val_metrics['mae_angle']:.3f} / {val_metrics['mae_speed']:.3f} / {val_mae_mean:.3f} | "
            f"Val RMSE (ang/spd/mean): {val_metrics['rmse_angle']:.3f} / {val_metrics['rmse_speed']:.3f} / {val_metrics['rmse_mean']:.3f} | "
            f"Test MAE (ang/spd/mean): {test_metrics['mae_angle']:.3f} / {test_metrics['mae_speed']:.3f} / {test_mae_mean:.3f} | "
            f"Test RMSE (ang/spd/mean): {test_metrics['rmse_angle']:.3f} / {test_metrics['rmse_speed']:.3f} / {test_metrics['rmse_mean']:.3f}"
        )

    # Save best-by-VAL MAE
    if val_mae_mean < best_val_mae:
        best_val_mae = val_mae_mean
        torch.save(model.state_dict(), best_val_path)
        print(f"New best-by-VAL saved: {best_val_path} (Val MAE mean: {best_val_mae:.3f})")

    # Save best-by-TEST MAE (for reporting only)
    if test_mae_mean < best_test_mae:
        best_test_mae = test_mae_mean
        torch.save(model.state_dict(), best_test_path)
        print(f"New best-by-TEST saved: {best_test_path} (Test MAE mean: {best_test_mae:.3f})")

    # Optional per-epoch checkpoint
    torch.save(model.state_dict(), save_every_epoch_path.format(epoch+1))

print("Training completed.")
print(f"Best-by-VAL checkpoint:  {best_val_path}  (Val MAE mean:  {best_val_mae:.3f})")
print(f"Best-by-TEST checkpoint: {best_test_path} (Test MAE mean: {best_test_mae:.3f})")
```

```python
# eval_test_deg_mapping.py
import os
import math
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import pandas as pd

# -----------------------------
# Config
# -----------------------------
BEST_VAL_PATH = "best_regressor_val.pth"   # change if needed
OUT_DIR = "eval_test_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# Dataset normalization ranges (from your Dataset)
ANGLE_MIN_RAW, ANGLE_MAX_RAW = 1.0, 7.0        # raw "angle index" in [1..7]
SPEED_MIN_RAW, SPEED_MAX_RAW = 9.7, 16.6       # your Excel-derived speeds

# If your model expects 3-channel inputs (from grayscale), set True
EXPAND_TO_3C = False

device = 'cuda' if torch.cuda.is_available() else 'cpu'
device_type = 'cuda' if device == 'cuda' else 'cpu'

# -----------------------------
# Import your model & dataloader
# -----------------------------
# from your_model_file import CRNNRegressor
# from your_data_file import test_dataloader
# ^ Ensure these are properly available in your environment.

# Placeholder import pattern; replace with your actual import.
# For example:
# from models.crnn import CRNNRegressor
# from data.loaders import test_dataloader

# -----------------------------
# Angle mapping helpers
# -----------------------------
def denorm_angle_index(angle_norm: np.ndarray) -> np.ndarray:
    """
    angle_norm in [0,1] -> angle_index in [1..7] (float).
    """
    return angle_norm * (ANGLE_MAX_RAW - ANGLE_MIN_RAW) + ANGLE_MIN_RAW

def angle_index_to_degrees(angle_idx: np.ndarray) -> np.ndarray:
    """
    Map angle index 1..7 to degrees 0..90 linearly.
    1 -> 0°, 7 -> 90°. That’s:
      deg = (angle_idx - 1) / (7 - 1) * 90 = 15 * (angle_idx - 1)
    """
    return 15.0 * (angle_idx - 1.0)

def denorm_speed(speed_norm: np.ndarray) -> np.ndarray:
    """
    speed_norm in [0,1] -> original speed units.
    """
    return speed_norm * (SPEED_MAX_RAW - SPEED_MIN_RAW) + SPEED_MIN_RAW

# -----------------------------
# Metrics
# -----------------------------
def mae_rmse(y_true: np.ndarray, y_pred: np.ndarray):
    err = y_pred - y_true
    if err.size == 0:
        return 0.0, 0.0
    mae = float(np.mean(np.abs(err)))
    rmse = float(math.sqrt(np.mean(err ** 2)))
    return mae, rmse

# -----------------------------
# Evaluation
# -----------------------------
@torch.no_grad()
def run_test(test_loader, model, use_amp=True):
    model.eval()

    folders = []
    angle_true_deg, angle_pred_deg = [], []
    speed_true, speed_pred = [], []

    with torch.autocast(device_type=device_type, enabled=(use_amp and device == 'cuda')):
        for batch in test_loader:
            # Dataset returns: frames [B,T,1,224,224], target [B,2], folder
            if len(batch) == 3:
                data, target, folder = batch
            else:
                data, target = batch
                folder = ["(unknown)"] * data.size(0)

            data = data.to(device, non_blocking=(device == 'cuda'))
            target = target.to(device, non_blocking=(device == 'cuda'))

            if EXPAND_TO_3C and data.size(2) == 1:
                # [B, T, 1, H, W] -> [B, T, 3, H, W]
                data = data.repeat(1, 1, 3, 1, 1)

            pred = model(data)  # [B, 2] in normalized space

            # Split normalized tensors -> numpy
            t_angle_norm = target[:, 0].detach().cpu().numpy()
            t_speed_norm = target[:, 1].detach().cpu().numpy()
            p_angle_norm = pred[:, 0].detach().cpu().numpy()
            p_speed_norm = pred[:, 1].detach().cpu().numpy()

            # De-normalize angles to index [1..7], then map to degrees [0..90]
            t_angle_idx = denorm_angle_index(t_angle_norm)
            p_angle_idx = denorm_angle_index(p_angle_norm)
            t_angle_deg = angle_index_to_degrees(t_angle_idx)
            p_angle_deg = angle_index_to_degrees(p_angle_idx)

            # De-normalize speed to original units
            t_speed = denorm_speed(t_speed_norm)
            p_speed = denorm_speed(p_speed_norm)

            angle_true_deg.append(t_angle_deg)
            angle_pred_deg.append(p_angle_deg)
            speed_true.append(t_speed)
            speed_pred.append(p_speed)

            # Collect folder identifiers
            if isinstance(folder, (list, tuple)):
                folders.extend(list(folder))
            else:
                try:
                    folders.extend(list(folder))
                except Exception:
                    folders.extend(["(unknown)"] * len(t_angle_deg))

    # Concatenate
    angle_true_deg = np.concatenate(angle_true_deg, axis=0) if angle_true_deg else np.array([])
    angle_pred_deg = np.concatenate(angle_pred_deg, axis=0) if angle_pred_deg else np.array([])
    speed_true = np.concatenate(speed_true, axis=0) if speed_true else np.array([])
    speed_pred = np.concatenate(speed_pred, axis=0) if speed_pred else np.array([])

    return folders, angle_true_deg, angle_pred_deg, speed_true, speed_pred

# -----------------------------
# Main
# -----------------------------
def main():
    # Instantiate and load model
    model = CRNNRegressor()
    model = nn.DataParallel(model).to(device)

    # Robustly load state dict (with/without DataParallel prefix)
    ckpt = torch.load(BEST_VAL_PATH, map_location=device)
    try:
        model.load_state_dict(ckpt, strict=True)
    except RuntimeError:
        from collections import OrderedDict
        new_state = OrderedDict((k.replace("module.", "", 1), v) for k, v in ckpt.items())
        try:
            model.module.load_state_dict(new_state, strict=True)
        except Exception:
            model.load_state_dict(new_state, strict=True)

    # Run test
    use_amp = (device == 'cuda')
    folders, a_t_deg, a_p_deg, s_t, s_p = run_test(test_dataloader, model, use_amp=use_amp)

    # Metrics (angle in degrees)
    angle_mae, angle_rmse = mae_rmse(a_t_deg, a_p_deg)
    speed_mae, speed_rmse = mae_rmse(s_t, s_p)
    mean_mae  = 0.5 * (angle_mae + speed_mae)
    mean_rmse = 0.5 * (angle_rmse + speed_rmse)

    print("\n=== Test metrics (de-normalized) ===")
    print(f"Angle (deg): MAE={angle_mae:.3f}, RMSE={angle_rmse:.3f}")
    print(f"Speed      : MAE={speed_mae:.3f}, RMSE={speed_rmse:.3f}")
    print(f"Mean       : MAE={mean_mae:.3f}, RMSE={mean_rmse:.3f}")

    # Save CSV
    n = min(len(folders), len(a_t_deg))
    df = pd.DataFrame({
        "folder": folders[:n],
        "angle_true_deg": a_t_deg[:n],
        "angle_pred_deg": a_p_deg[:n],
        "angle_err_deg":  (a_p_deg[:n] - a_t_deg[:n]),
        "speed_true": s_t[:n],
        "speed_pred": s_p[:n],
        "speed_err":  (s_p[:n] - s_t[:n]),
    })
    csv_path = os.path.join(OUT_DIR, "test_predictions_deg.csv")
    df.to_csv(csv_path, index=False)
    print(f"Saved per-sample predictions to: {csv_path}")

    # -------------------------
    # Plots
    # -------------------------
    # 1) Angle degrees: pred vs true
    if a_t_deg.size:
        plt.figure()
        plt.scatter(a_t_deg, a_p_deg, s=10, alpha=0.6)
        mn = float(min(a_t_deg.min(), a_p_deg.min()))
        mx = float(max(a_t_deg.max(), a_p_deg.max()))
        plt.plot([mn, mx], [mn, mx])
        plt.xlabel("Angle (true, deg)")
        plt.ylabel("Angle (pred, deg)")
        plt.title(f"Angle: Pred vs True (MAE={angle_mae:.2f}°, RMSE={angle_rmse:.2f}°)")
        plt.tight_layout()
        out_path = os.path.join(OUT_DIR, "angle_deg_pred_vs_true.png")
        plt.savefig(out_path, dpi=160)
        plt.close()
        print("Saved:", out_path)

        # 2) Angle residuals (deg)
        plt.figure()
        plt.hist(a_p_deg - a_t_deg, bins=40)
        plt.xlabel("Angle Error (deg, pred - true)")
        plt.ylabel("Count")
        plt.title("Angle Residuals (deg)")
        plt.tight_layout()
        out_path = os.path.join(OUT_DIR, "angle_deg_residuals_hist.png")
        plt.savefig(out_path, dpi=160)
        plt.close()
        print("Saved:", out_path)

    # 3) Speed: pred vs true
    if s_t.size:
        plt.figure()
        plt.scatter(s_t, s_p, s=10, alpha=0.6)
        mn = float(min(s_t.min(), s_p.min()))
        mx = float(max(s_t.max(), s_p.max()))
        plt.plot([mn, mx], [mn, mx])
        plt.xlabel("Speed (true)")
        plt.ylabel("Speed (pred)")
        plt.title(f"Speed: Pred vs True (MAE={speed_mae:.2f}, RMSE={speed_rmse:.2f})")
        plt.tight_layout()
        out_path = os.path.join(OUT_DIR, "speed_pred_vs_true.png")
        plt.savefig(out_path, dpi=160)
        plt.close()
        print("Saved:", out_path)

        # 4) Speed residuals
        plt.figure()
        plt.hist(s_p - s_t, bins=40)
        plt.xlabel("Speed Error (pred - true)")
        plt.ylabel("Count")
        plt.title("Speed Residuals")
        plt.tight_layout()
        out_path = os.path.join(OUT_DIR, "speed_residuals_hist.png")
        plt.savefig(out_path, dpi=160)
        plt.close()
        print("Saved:", out_path)



if __name__ == "__main__":
    # Make sure CRNNRegressor and test_dataloader are importable above.
    main()
```

```python
# -------------------------
# 2D visualizations
# -------------------------
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
df = pd.read_csv("eval_test_outputs/test_predictions_deg.csv")
a_t_deg = df["angle_true_deg"].to_numpy()
a_p_deg = df["angle_pred_deg"].to_numpy()
s_t     = df["speed_true"].to_numpy()
s_p     = df["speed_pred"].to_numpy()
OUT_DIR = "eval_test_outputs"



# Helper to save hexbin with diagonal
def plot_hexbin_true_pred(y_true, y_pred, xlabel, ylabel, title, fname, gridsize=35):
    if y_true.size == 0:
        return
    plt.figure()
    hb = plt.hexbin(y_true, y_pred, gridsize=gridsize, mincnt=1, bins='log')
    mn = float(min(y_true.min(), y_pred.min()))
    mx = float(max(y_true.max(), y_pred.max()))
    plt.plot([mn, mx], [mn, mx], linewidth=1)
    cb = plt.colorbar(hb)
    cb.set_label('log10(count)')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.tight_layout()
    path = os.path.join(OUT_DIR, fname)
    plt.savefig(path, dpi=180)
    plt.close()
    print("Saved:", path)

# 1) Angle calibration: true vs pred (hexbin)
plot_hexbin_true_pred(
    a_t_deg, a_p_deg,
    xlabel="Angle True (deg)",
    ylabel="Angle Pred (deg)",
    title="Angle Calibration (Hexbin)",
    fname="angle_true_vs_pred_hexbin.png",
)

# 2) Speed calibration: true vs pred (hexbin)
plot_hexbin_true_pred(
    s_t, s_p,
    xlabel="Speed True",
    ylabel="Speed Pred",
    title="Speed Calibration (Hexbin)",
    fname="speed_true_vs_pred_hexbin.png",
)

# 3) Joint 2D: (true angle deg, true speed) colored by |error|
#    Useful to see where the model struggles in the state space.
if a_t_deg.size and s_t.size:
    abs_err_angle = np.abs(a_p_deg - a_t_deg)
    abs_err_speed = np.abs(s_p - s_t)

    plt.figure()
    sc = plt.scatter(a_t_deg, s_t, c=abs_err_angle, s=12, alpha=0.8)
    cb = plt.colorbar(sc)
    cb.set_label("|Angle Error| (deg)")
    plt.xlabel("Angle True (deg)")
    plt.ylabel("Speed True")
    plt.title("Error Field: |Angle Error| over (Angle, Speed)")
    plt.tight_layout()
    path = os.path.join(OUT_DIR, "error_field_angle_over_angle_speed.png")
    plt.savefig(path, dpi=180)
    plt.close()
    print("Saved:", path)

    plt.figure()
    sc = plt.scatter(a_t_deg, s_t, c=abs_err_speed, s=12, alpha=0.8)
    cb = plt.colorbar(sc)
    cb.set_label("|Speed Error|")
    plt.xlabel("Angle True (deg)")
    plt.ylabel("Speed True")
    plt.title("Error Field: |Speed Error| over (Angle, Speed)")
    plt.tight_layout()
    path = os.path.join(OUT_DIR, "error_field_speed_over_angle_speed.png")
    plt.savefig(path, dpi=180)
    plt.close()
    print("Saved:", path)

# 4) Binned heatmap of mean absolute error over (angle_true_deg, speed_true)
#    This aggregates errors per cell for a smoother, grid-like view.
def binned_mae_heatmap(x_true, y_true, err, xbins, ybins, xlabel, ylabel, title, fname):
    if x_true.size == 0:
        return
    H_sum = np.zeros((len(xbins)-1, len(ybins)-1), dtype=float)
    H_cnt = np.zeros_like(H_sum)

    xi = np.digitize(x_true, xbins) - 1  # 0..nbins-1
    yi = np.digitize(y_true, ybins) - 1
    mask = (xi >= 0) & (xi < H_sum.shape[0]) & (yi >= 0) & (yi < H_sum.shape[1])

    for i, j, e in zip(xi[mask], yi[mask], err[mask]):
        H_sum[i, j] += abs(e)
        H_cnt[i, j] += 1

    with np.errstate(invalid='ignore', divide='ignore'):
        H_mae = np.where(H_cnt > 0, H_sum / H_cnt, np.nan)

    # Plot
    plt.figure()
    # extent expects [xmin, xmax, ymin, ymax]
    extent = [xbins[0], xbins[-1], ybins[0], ybins[-1]]
    im = plt.imshow(H_mae.T, origin='lower', extent=extent, aspect='auto')
    cb = plt.colorbar(im)
    cb.set_label("Mean |Error|")
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.tight_layout()
    path = os.path.join(OUT_DIR, fname)
    plt.savefig(path, dpi=180)
    plt.close()
    print("Saved:", path)

    

# Choose bin edges (tune to your data spread)
if a_t_deg.size and s_t.size:
    angle_bins = np.linspace(0, 90, 19)  # 18 bins of 5°
    speed_bins = np.linspace(float(s_t.min()), float(s_t.max()), 16)  # 15 bins

    binned_mae_heatmap(
        a_t_deg, s_t, (a_p_deg - a_t_deg),
        angle_bins, speed_bins,
        xlabel="Angle True (deg)",
        ylabel="Speed True",
        title="Binned MAE (Angle) over (Angle True, Speed True)",
        fname="binned_mae_angle_over_angle_speed.png",
    )
    binned_mae_heatmap(
        a_t_deg, s_t, (s_p - s_t),
        angle_bins, speed_bins,
        xlabel="Angle True (deg)",
        ylabel="Speed True",
        title="Binned MAE (Speed) over (Angle True, Speed True)",
        fname="binned_mae_speed_over_angle_speed.png",
    )
```

```python
import os
import math
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import pandas as pd

# ---------- config ----------
BEST_VAL_PATH = "best_regressor_val.pth"   # change if your filename differs
OUT_DIR = "eval_test_outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# Match your dataset's normalization ranges
ANGLE_MIN, ANGLE_MAX = 1.0, 7.0
SPEED_MIN, SPEED_MAX = 9.7, 16.6

# If your model expects 3-channel inputs from grayscale, set True
EXPAND_TO_3C = False

device = 'cuda' if torch.cuda.is_available() else 'cpu'
device_type = 'cuda' if device == 'cuda' else 'cpu'

# ---------- model ----------
# from your_model_file import CRNNRegressor   # uncomment & adjust if needed
model = CRNNRegressor()
model = nn.DataParallel(model).to(device)

# Robust load for weights saved with/without DataParallel
ckpt = torch.load(BEST_VAL_PATH, map_location=device)
try:
    model.load_state_dict(ckpt, strict=True)
except RuntimeError:
    # Strip "module." prefix if necessary
    from collections import OrderedDict
    new_state = OrderedDict()
    for k, v in ckpt.items():
        new_state[k.replace("module.", "", 1)] = v
    try:
        model.module.load_state_dict(new_state, strict=True)
    except Exception:
        model.load_state_dict(new_state, strict=True)

model.eval()

# ---------- helpers ----------
def denorm_angle(x):
    return x * (ANGLE_MAX - ANGLE_MIN) + ANGLE_MIN

def denorm_speed(x):
    return x * (SPEED_MAX - SPEED_MIN) + SPEED_MIN

@torch.no_grad()
def run_test(test_loader, use_amp=True):
    # storage
    folders = []
    angle_true, angle_pred = [], []
    speed_true, speed_pred = [], []

    with torch.autocast(device_type=device_type, enabled=(use_amp and device=='cuda')):
        for batch in test_loader:
            # Dataset returns: frames [B,T,1,224,224], target [B,2], folder (list/tuple)
            if len(batch) == 3:
                data, target, folder = batch
            else:
                data, target = batch
                folder = ["(unknown)"] * data.size(0)

            data = data.to(device, non_blocking=(device=='cuda'))
            target = target.to(device, non_blocking=(device=='cuda'))

            if EXPAND_TO_3C and data.size(2) == 1:
                # [B, T, 1, H, W] -> [B, T, 3, H, W]
                data = data.repeat(1, 1, 3, 1, 1)

            # forward
            pred = model(data)  # [B, 2] in normalized space

            # split, CPU numpy
            t_angle_norm = target[:, 0].detach().cpu().numpy()
            t_speed_norm = target[:, 1].detach().cpu().numpy()
            p_angle_norm = pred[:, 0].detach().cpu().numpy()
            p_speed_norm = pred[:, 1].detach().cpu().numpy()

            # de-normalize to original units
            t_angle = denorm_angle(t_angle_norm)
            t_speed = denorm_speed(t_speed_norm)
            p_angle = denorm_angle(p_angle_norm)
            p_speed = denorm_speed(p_speed_norm)

            angle_true.append(t_angle)
            angle_pred.append(p_angle)
            speed_true.append(t_speed)
            speed_pred.append(p_speed)

            # folders (list of strings)
            if isinstance(folder, (list, tuple)):
                folders.extend(list(folder))
            else:
                # sometimes collate packs as list of strings already
                try:
                    folders.extend(list(folder))
                except Exception:
                    folders.extend(["(unknown)"] * len(t_angle))

    angle_true = np.concatenate(angle_true, axis=0) if len(angle_true) else np.array([])
    angle_pred = np.concatenate(angle_pred, axis=0) if len(angle_pred) else np.array([])
    speed_true = np.concatenate(speed_true, axis=0) if len(speed_true) else np.array([])
    speed_pred = np.concatenate(speed_pred, axis=0) if len(speed_pred) else np.array([])

    return folders, angle_true, angle_pred, speed_true, speed_pred

def mae_rmse(y_true, y_pred):
    err = y_pred - y_true
    mae = np.mean(np.abs(err)) if err.size else 0.0
    rmse = math.sqrt(np.mean(err**2)) if err.size else 0.0
    return mae, rmse

# ---------- run ----------
use_amp = (device == 'cuda')
folders, a_t, a_p, s_t, s_p = run_test(test_dataloader, use_amp=use_amp)

# metrics
angle_mae, angle_rmse = mae_rmse(a_t, a_p)
speed_mae, speed_rmse = mae_rmse(s_t, s_p)
mean_mae  = 0.5 * (angle_mae + speed_mae)
mean_rmse = 0.5 * (angle_rmse + speed_rmse)

print("\n=== Test metrics (de-normalized) ===")
print(f"Angle  : MAE={angle_mae:.3f}, RMSE={angle_rmse:.3f}")
print(f"Speed  : MAE={speed_mae:.3f}, RMSE={speed_rmse:.3f}")
print(f"Mean   : MAE={mean_mae:.3f}, RMSE={mean_rmse:.3f}")

# ---------- save CSV ----------
df = pd.DataFrame({
    "folder": folders[:len(a_t)],
    "angle_true": a_t, "angle_pred": a_p, "angle_err": (a_p - a_t),
    "speed_true": s_t, "speed_pred": s_p, "speed_err": (s_p - s_t),
})
csv_path = os.path.join(OUT_DIR, "test_predictions.csv")
df.to_csv(csv_path, index=False)
print(f"Saved per-sample predictions to: {csv_path}")

# ---------- plots ----------
# 1) Angle: predicted vs true
plt.figure()
plt.scatter(a_t, a_p, s=10, alpha=0.6)
mn = min(a_t.min(), a_p.min()) if len(a_t) else 0
mx = max(a_t.max(), a_p.max()) if len(a_t) else 1
plt.plot([mn, mx], [mn, mx])
plt.xlabel("Angle (true)")
plt.ylabel("Angle (pred)")
plt.title(f"Angle: Pred vs True (MAE={angle_mae:.2f}, RMSE={angle_rmse:.2f})")
plt.tight_layout()
angle_scatter_path = os.path.join(OUT_DIR, "angle_pred_vs_true.png")
plt.savefig(angle_scatter_path, dpi=160)
plt.close()

# 2) Speed: predicted vs true
plt.figure()
plt.scatter(s_t, s_p, s=10, alpha=0.6)
mn = min(s_t.min(), s_p.min()) if len(s_t) else 0
mx = max(s_t.max(), s_p.max()) if len(s_t) else 1
plt.plot([mn, mx], [mn, mx])
plt.xlabel("Speed (true)")
plt.ylabel("Speed (pred)")
plt.title(f"Speed: Pred vs True (MAE={speed_mae:.2f}, RMSE={speed_rmse:.2f})")
plt.tight_layout()
speed_scatter_path = os.path.join(OUT_DIR, "speed_pred_vs_true.png")
plt.savefig(speed_scatter_path, dpi=160)
plt.close()

# 3) Residual histograms
plt.figure()
plt.hist(a_p - a_t, bins=40)
plt.xlabel("Angle Error (pred - true)")
plt.ylabel("Count")
plt.title("Angle Residuals")
plt.tight_layout()
angle_hist_path = os.path.join(OUT_DIR, "angle_residuals_hist.png")
plt.savefig(angle_hist_path, dpi=160)
plt.close()

plt.figure()
plt.hist(s_p - s_t, bins=40)
plt.xlabel("Speed Error (pred - true)")
plt.ylabel("Count")
plt.title("Speed Residuals")
plt.tight_layout()
speed_hist_path = os.path.join(OUT_DIR, "speed_residuals_hist.png")
plt.savefig(speed_hist_path, dpi=160)
plt.close()

print("Saved plots:")
print(" -", angle_scatter_path)
print(" -", speed_scatter_path)
print(" -", angle_hist_path)
print(" -", speed_hist_path)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import math

# -----------------------------
# Model: use your regressor
# -----------------------------
# If your CRNNRegressor's first conv is set to C=1, keep as-is.
# If you kept it at 3 channels, set EXPAND_TO_3C=True below to repeat grayscale.
#from your_model_file import CRNNRegressor  # adjust import

EXPAND_TO_3C = False  # set True ONLY if model expects 3-channel input

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = CRNNRegressor()
model = nn.DataParallel(model)             # OK on single GPU/CPU too
model = model.to(device)

# Optional speed tweak on GPU
if device == 'cuda':
    torch.backends.cudnn.benchmark = True

# -----------------------------
# Metrics (validation/test)
# -----------------------------
@torch.no_grad()
def evaluate_regression(dataloader, model, device, use_amp: bool):
    model.eval()
    sum_mae_angle = 0.0
    sum_mae_speed = 0.0
    sum_mse_angle = 0.0
    sum_mse_speed = 0.0
    n = 0

    with torch.amp.autocast(device, enabled=use_amp):
        for data, target, *_ in dataloader:
            # data: [B, T, 1, 224, 224]; target: [B, 2] = [angle, speed]
            data = data.to(device, non_blocking=True) if device == 'cuda' else data.to(device)
            target = target.to(device, non_blocking=True) if device == 'cuda' else target.to(device)

            if EXPAND_TO_3C and data.size(2) == 1:
                # Repeat grayscale channel to RGB for models expecting 3 channels
                data = data.repeat(1, 1, 3, 1, 1)

            pred = model(data)  # [B, 2]
            diff = pred - target  # [B, 2]

            mae = diff.abs()     # [B, 2]
            mse = diff.pow(2)    # [B, 2]

            sum_mae_angle += mae[:, 0].sum().item()
            sum_mae_speed += mae[:, 1].sum().item()
            sum_mse_angle += mse[:, 0].sum().item()
            sum_mse_speed += mse[:, 1].sum().item()
            n += target.size(0)

    model.train()
    if n == 0:
        return {
            "mae_angle": 0.0, "mae_speed": 0.0, "mae_mean": 0.0,
            "rmse_angle": 0.0, "rmse_speed": 0.0, "rmse_mean": 0.0
        }

    mae_angle = sum_mae_angle / n
    mae_speed = sum_mae_speed / n
    rmse_angle = math.sqrt(sum_mse_angle / n)
    rmse_speed = math.sqrt(sum_mse_speed / n)
    return {
        "mae_angle": mae_angle,
        "mae_speed": mae_speed,
        "mae_mean": 0.5 * (mae_angle + mae_speed),
        "rmse_angle": rmse_angle,
        "rmse_speed": rmse_speed,
        "rmse_mean": 0.5 * (rmse_angle + rmse_speed),
    }

# -----------------------------
# Loss / optimizer / AMP
# -----------------------------
# SmoothL1 is robust. You can switch to MSELoss() if you prefer.
criterion = nn.SmoothL1Loss(beta=0.1)  # Huber loss
# criterion = nn.MSELoss()

optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)

use_amp = (device == 'cuda')
scaler = torch.amp.GradScaler(device) if use_amp else None

best_val_mae = float('inf')
best_model_path = 'best_regressor.pth'
save_every_epoch_path = 'model_epoch{}.pth'  # optional per-epoch save

# -----------------------------
# Training loop
# -----------------------------
num_epochs = 200  # start smaller than 800 for regression; adjust as needed
log_every = 1
grad_clip = 1.0   # set None to disable

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    batch_count = 0

    for data, target, *_ in train_dataloader:
        data = data.to(device, non_blocking=True) if device == 'cuda' else data.to(device)
        target = target.to(device, non_blocking=True) if device == 'cuda' else target.to(device)

        if EXPAND_TO_3C and data.size(2) == 1:
            data = data.repeat(1, 1, 3, 1, 1)

        optimizer.zero_grad(set_to_none=True)

        with torch.amp.autocast(device, enabled=use_amp):
            pred = model(data)          # [B, 2]
            loss = criterion(pred, target)

        if use_amp:
            scaler.scale(loss).backward()
            if grad_clip is not None:
                scaler.unscale_(optimizer)
                nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
            scaler.step(optimizer)
            scaler.update()
        else:
            loss.backward()
            if grad_clip is not None:
                nn.utils.clip_grad_norm_(model.parameters(), grad_clip)
            optimizer.step()

        running_loss += loss.item()
        batch_count += 1

    # ---- end epoch: evaluate ----
    train_loss = running_loss / max(1, batch_count)
    val_metrics = evaluate_regression(val_dataloader, model, device, use_amp)
    val_mae = val_metrics["rmse_mean"]

    if (epoch + 1) % log_every == 0:
        print(
            f"Epoch [{epoch+1}/{num_epochs}] "
            f"TrainLoss: {train_loss:.4f} | "
            f"Val MAE (angle/speed/mean): "
            f"{val_metrics['mae_angle']:.3f} / {val_metrics['mae_speed']:.3f} / {val_metrics['mae_mean']:.3f} | "
            f"Val RMSE (angle/speed/mean): "
            f"{val_metrics['rmse_angle']:.3f} / {val_metrics['rmse_speed']:.3f} / {val_metrics['rmse_mean']:.3f}"
        )

    # Save best by lowest validation MAE
    if val_mae < best_val_mae:
        best_val_mae = val_mae
        torch.save(model.state_dict(), best_model_path)
        print(f"New best saved: {best_model_path} (Val MAE mean: {best_val_mae:.3f})")

    # Optional per-epoch checkpoint
    # torch.save(model.state_dict(), save_every_epoch_path.format(epoch+1))

print("Training completed.")
print(f"Best model (lowest Val MAE) saved at: {best_model_path}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
import torch

#model = SingleChannelVivitModel(config)
#device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
#model = nn.DataParallel(model)
#model = model.to(device)

#best_model_path = 'best_vivit_model100_diff_3.pth'

def get_all_predictions_with_paths(dataloader, model, device, save_to_file='predictions_with_paths.txt'):
    model.eval()
    all_preds = []
    all_labels = []
    all_paths = []
    
    with torch.no_grad():
        with open(save_to_file, 'w') as f:
            for data, labels, paths in dataloader:
                data = data.to(device)
                labels = labels.to(device)
                outputs = model(pixel_values=data)
                _, predicted = torch.max(outputs, 1)

                all_preds.extend(predicted.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
                all_paths.extend(paths)

                # Write the file paths and predicted labels to the text file
                for path, pred_label in zip(paths, predicted.cpu().numpy()):
                    f.write(f'{path}, {pred_label}\n')

    return np.array(all_preds), np.array(all_labels), all_paths


model.load_state_dict(torch.load(best_model_path))
# 获取所有预测、真实标签以及文件路径
#pred_labels, true_labels, file_paths = get_all_predictions_with_paths(val_dataloader, model, device)
pred_labels, true_labels = get_all_predictions(val_dataloader, model, device)

accuracy = accuracy_score(true_labels, pred_labels)
# 计算混淆矩阵
conf_matrix = confusion_matrix(true_labels, pred_labels)

# 获取类别名称
#class_names = ['bishop', 'king', 'knight', 'pawn', 'queen', 'rook']



# 将混淆矩阵转换为百分比
conf_matrix_percent = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis] * 100

# 可视化百分比混淆矩阵
plt.figure(figsize=(8, 8))

sns.heatmap(conf_matrix_percent, annot=True, fmt='.2f', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names,
            annot_kws={"size": 16},  cbar=False)  # 调整注释字体大小
plt.xlabel('Predicted Label', fontsize=16)  # 调整x轴标签字体大小
plt.ylabel('True Label', fontsize=16)  # 调整y轴标签字体大小
plt.xticks(fontsize=16)  # 调整x轴刻度字体大小
plt.yticks(fontsize=16)  # 调整y轴刻度字体大小
plt.title('Confusion Matrix (in %) - Overall Accuracy: {:.2f}'.format(accuracy * 100), fontsize=15)
plt.tight_layout()

#pdf_name = 'AT_Diff_confusion_matrix.pdf'
#plt.savefig(pdf_name, format='pdf')
plt.show()

print(f'Predictions and file paths saved to {save_to_file}')
```
