# Chess validation/test evaluation

This notebook evaluates the CRNN model on the chess validation/test dataset used in the manuscript.

**Expected data layout**

```text
DATA_ROOT/
  ClassName_1/
    sample_001/
      frame_0001.jpg
      frame_0002.jpg
      ...
  ClassName_2/
    sample_002/
      frame_0001.jpg
      ...
```

**Outputs**

The notebook saves the following files under `OUTPUT_DIR`:

- `predictions.csv`: per-sample ground truth and predicted labels
- `metrics.json`: overall accuracy and class names
- `confusion_matrix_counts.csv`: raw confusion matrix
- `confusion_matrix_percent.csv`: row-normalized confusion matrix in percent
- `confusion_matrix_percent.png`: figure used for reporting/inspection

````python
# Configuration

from pathlib import Path

# Update these paths after downloading the validation/test data and model checkpoint.
DATA_ROOT = Path("./FibTac_data/Chess_validation")
CHECKPOINT_PATH = Path(r"./checkpoints/crnn_chess_weights.pth")
OUTPUT_DIR = Path("./results/chess")

MAX_FRAMES = 150
FRAME_STRIDE = 2
START_INDEX = 0
BATCH_SIZE = 1
NUM_WORKERS = 0  # Keep 0 when tensors are moved to DEVICE inside the dataset.

SEED = 42
````

````python
# Imports and reproducibility settings

import json
import random
from pathlib import Path

import cv2
import kornia.augmentation as K
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from kornia.augmentation import VideoSequential
from sklearn.metrics import accuracy_score, confusion_matrix
from torch.utils.data import DataLoader, Dataset
from torchvision import models

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print(f"Using device: {DEVICE}")
````

````python
# Dataset and preprocessing utilities

class ClassDataset(Dataset):
    """
    Dataset for folders of image-frame sequences.

    Each sample is one instance folder containing .jpg frames. Class labels are
    inferred from the parent folder name. For evaluation, frames are read in a
    deterministic order, every FRAME_STRIDE-th frame is used, and at most
    MAX_FRAMES frames are retained. Frames are processed with the same
    validation/test transform pattern as the original task code:

    CenterCrop(480 x 480) -> Resize(224 x 224) -> Normalize(mean=0.5, std=0.5)
    """

    def __init__(
        self,
        root_dir,
        device,
        max_frames=150,
        frame_stride=2,
        start_index=0,
    ):
        self.root_dir = Path(root_dir)
        self.device = torch.device(device)
        self.max_frames = int(max_frames)
        self.frame_stride = int(frame_stride)
        self.start_index = int(start_index)

        self.class_map = {}
        self.paths = []
        self.labels = []

        self.transform = VideoSequential(
            K.CenterCrop((480, 480)),
            K.Resize((224, 224)),
            K.Normalize(mean=torch.tensor([0.5]), std=torch.tensor([0.5])),
        ).to(self.device)

        self._index_folders()

    def _index_folders(self):
        if not self.root_dir.exists():
            raise FileNotFoundError(f"DATA_ROOT does not exist: {self.root_dir}")

        label_map = {}
        next_label = 0

        for class_folder in sorted(self.root_dir.iterdir()):
            if not class_folder.is_dir() or class_folder.name.startswith("."):
                continue

            if class_folder.name not in label_map:
                label_map[class_folder.name] = next_label
                self.class_map[next_label] = class_folder.name
                next_label += 1

            label = label_map[class_folder.name]

            for instance_folder in sorted(class_folder.iterdir()):
                if instance_folder.is_dir() and not instance_folder.name.startswith("."):
                    self.paths.append(instance_folder)
                    self.labels.append(label)

        if len(self.paths) == 0:
            raise ValueError(f"No sample folders were found under DATA_ROOT: {self.root_dir}")

    def _read_frames(self, folder_path: Path) -> np.ndarray:
        frame_paths = sorted(folder_path.glob("*.jpg"))
        frames = []

        selected_paths = frame_paths[self.start_index :: self.frame_stride]
        for frame_path in selected_paths:
            img = cv2.imread(str(frame_path), cv2.IMREAD_GRAYSCALE)
            if img is None:
                continue

            frames.append(img)

            if len(frames) >= self.max_frames:
                break

        if len(frames) == 0:
            print(f"[WARNING] No valid frames in: {folder_path}")
            frames = [np.zeros((480, 480), dtype=np.float32)]

        return np.array(frames, dtype=np.float32)

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        folder = self.paths[idx]
        label = int(self.labels[idx])

        frames = self._read_frames(folder)

        frames_tensor = torch.stack(
            [torch.tensor(frame, dtype=torch.float32).unsqueeze(0) for frame in frames]
        )  # [time, channel, height, width]
        frames_tensor = frames_tensor / 255.0
        frames_tensor = frames_tensor.to(self.device)

        # Kornia VideoSequential expects [batch, time, channel, height, width].
        frames_tensor = frames_tensor.unsqueeze(0)
        frames_tensor = self.transform(frames_tensor)
        frames_tensor = frames_tensor.squeeze(0)

        return frames_tensor, torch.tensor(label, dtype=torch.long), str(folder)
````

````python
# Model definition

class CRNN(nn.Module):
    def __init__(self, num_classes=6):
        super().__init__()

        self.resnet = models.resnet18(weights=None)
        self.resnet.conv1 = nn.Conv2d(
            1,
            64,
            kernel_size=(7, 7),
            stride=(2, 2),
            padding=(3, 3),
            bias=False,
        )
        self.resnet = nn.Sequential(*list(self.resnet.children())[:-1])

        self.lstm = nn.LSTM(512, 128, batch_first=True)

        self.mlp2 = nn.Sequential(
            nn.Linear(128, 32),
            nn.ReLU(),
            nn.Linear(32, num_classes),
        )

    def forward(self, x):
        # Expected input: [batch, time, channel, height, width]
        batch_size, timesteps, channels, height, width = x.size()
        cnn_input = x.view(batch_size * timesteps, channels, height, width)

        features = self.resnet(cnn_input)
        features = features.view(batch_size, timesteps, -1)

        output, _ = self.lstm(features)
        output = self.mlp2(output[:, -1, :])
        return output
````

````python
# Create dataset and dataloader

test_dataset = ClassDataset(
    root_dir=DATA_ROOT,
    device=DEVICE,
    max_frames=MAX_FRAMES,
    frame_stride=FRAME_STRIDE,
    start_index=START_INDEX,
)

test_dataloader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    drop_last=False,
    num_workers=NUM_WORKERS,
)

class_names = [class_name for _, class_name in sorted(test_dataset.class_map.items())]

print(f"Number of samples: {len(test_dataset)}")
print(f"Classes: {class_names}")
````

````python
# Load checkpoint and run evaluation

def _strip_module_prefix(state_dict):
    """Remove 'module.' prefixes from DataParallel checkpoints when present."""
    return {
        key.replace("module.", "", 1) if key.startswith("module.") else key: value
        for key, value in state_dict.items()
    }


def load_state_dict_safely(model, checkpoint_path, device):
    checkpoint_path = Path(checkpoint_path)
    if not checkpoint_path.exists():
        raise FileNotFoundError(f"Checkpoint file does not exist: {checkpoint_path}")

    try:
        checkpoint = torch.load(checkpoint_path, map_location=device, weights_only=True)
    except TypeError:
        # For older PyTorch versions that do not support weights_only.
        checkpoint = torch.load(checkpoint_path, map_location=device)

    if isinstance(checkpoint, nn.Module):
        return checkpoint.to(device)

    if isinstance(checkpoint, dict) and "state_dict" in checkpoint:
        state_dict = checkpoint["state_dict"]
    elif isinstance(checkpoint, dict) and "model_state_dict" in checkpoint:
        state_dict = checkpoint["model_state_dict"]
    else:
        state_dict = checkpoint

    state_dict = _strip_module_prefix(state_dict)
    model.load_state_dict(state_dict)
    return model


def get_all_predictions(dataloader, model, device):
    model.eval()
    rows = []
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for data, labels, folders in dataloader:
            data = data.to(device)
            labels = labels.to(device)

            outputs = model(data)
            predicted = outputs.argmax(dim=1)

            for folder, true_label, pred_label in zip(folders, labels.cpu().numpy(), predicted.cpu().numpy()):
                rows.append(
                    {
                        "sample_path": folder,
                        "sample_id": Path(folder).name,
                        "true_label_id": int(true_label),
                        "true_label": class_names[int(true_label)],
                        "predicted_label_id": int(pred_label),
                        "predicted_label": class_names[int(pred_label)],
                    }
                )

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    return np.array(all_preds), np.array(all_labels), pd.DataFrame(rows)


num_classes = len(class_names)
model = CRNN(num_classes=num_classes)
model = load_state_dict_safely(model, CHECKPOINT_PATH, DEVICE)
model = model.to(DEVICE)

if torch.cuda.device_count() > 1 and not isinstance(model, nn.DataParallel):
    model = nn.DataParallel(model)

pred_labels, true_labels, prediction_table = get_all_predictions(test_dataloader, model, DEVICE)

accuracy = accuracy_score(true_labels, pred_labels)
conf_matrix = confusion_matrix(true_labels, pred_labels, labels=np.arange(num_classes))

row_sums = conf_matrix.sum(axis=1, keepdims=True)
conf_matrix_percent = np.divide(
    conf_matrix.astype(float),
    row_sums,
    out=np.zeros_like(conf_matrix, dtype=float),
    where=row_sums != 0,
) * 100

prediction_table.to_csv(OUTPUT_DIR / "predictions.csv", index=False)
pd.DataFrame(conf_matrix, index=class_names, columns=class_names).to_csv(
    OUTPUT_DIR / "confusion_matrix_counts.csv"
)
pd.DataFrame(conf_matrix_percent, index=class_names, columns=class_names).to_csv(
    OUTPUT_DIR / "confusion_matrix_percent.csv"
)

metrics = {
    "accuracy": float(accuracy),
    "accuracy_percent": float(accuracy * 100),
    "num_samples": int(len(true_labels)),
    "class_names": class_names,
    "data_root": str(DATA_ROOT),
    "checkpoint_path": str(CHECKPOINT_PATH),
    "max_frames": MAX_FRAMES,
    "frame_stride": FRAME_STRIDE,
    "start_index": START_INDEX,
}
with open(OUTPUT_DIR / "metrics.json", "w", encoding="utf-8") as f:
    json.dump(metrics, f, indent=2)

print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Saved results to: {OUTPUT_DIR}")
````

````python
# Plot and save the row-normalized confusion matrix

import seaborn as sns

figure_size = max(8, 0.8 * len(class_names))
plt.figure(figsize=(figure_size, figure_size))
sns.heatmap(
    conf_matrix_percent,
    annot=True,
    fmt=".2f",
    cmap="Blues",
    xticklabels=class_names,
    yticklabels=class_names,
    annot_kws={"size": 16},
    cbar=False,
)

plt.xlabel("Predicted label", fontsize=16)
plt.ylabel("True label", fontsize=16)
plt.xticks(fontsize=16, rotation=0, ha="center")
plt.yticks(fontsize=16, rotation=0)
plt.title(f"Confusion matrix (%) - Overall accuracy: {accuracy * 100:.2f}%", fontsize=15)
plt.tight_layout()

figure_path = OUTPUT_DIR / "confusion_matrix_percent.png"
plt.savefig(figure_path, dpi=300, bbox_inches="tight")
plt.show()

print(f"Saved figure to: {figure_path}")
````
